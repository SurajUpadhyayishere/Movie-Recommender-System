```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python

# Apply the function safely
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
             L.append(i['name'])
             counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return  L
```

```python
# Apply the function to the 'crew' column
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return "".join(y)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
